```python
my_catalog = dbutils.widgets.get("catalog")
my_schema = dbutils.widgets.get("schema")
```

```python
customerDf = (spark
              .read
              .format("csv")
              .option("header","true")
              .option("inferSchema","true")
              .load(f"/Volumes/{my_catalog}/{my_schema}/retail_customer_files/customers.csv"))
```

```python
customerDf.display()
```

```python
customerDf.write.mode("overwrite").saveAsTable(f"{my_catalog}.{my_schema}.customers_bronze")
```
