```python
%sql
create volume if not exists auto_loader_files 
```

```python
checkpoint_file_location = f'/Volumes/ecommerce/v01/auto_loader_files'

(spark
 .readStream
 .format("cloudFiles")
 .option("cloudFiles.format", "csv" )
 .option("header", "true")
 .option("sep" , ",")
 .option("cloudFiles.schemaLocation", checkpoint_file_location)
 .load("/Volumes/ecommerce/v01/csv_files_autoloader_source" )
 .writeStream
 .option("checkpointLocation", checkpoint_file_location)
 .trigger(once=True)
 .toTable("ecommerce.v01.python_csv_autoloader")
 )
```

```python
%sql
DESCRIBE EXTENDED ecommerce.v01.python_csv_autoloader
```
