```python
%sql
CREATE OR REPLACE TABLE ecommerce.retail_customer.customers_sales_silver AS 
SELECT 
c.customer_id,
c.customer_name,
c.state,
c.loyalty_segment,
c.units_purchased,
s.product_name,
s.order_date,
s.total_price
FROM ecommerce.retail_customer.customers_bronze c
JOIN ecommerce.retail_customer.sales_bronze s 
ON 
c.customer_id = s.customer_id
```

```python
df = spark.sql("SELECT * FROM ecommerce.retail_customer.customers_sales_silver")
duplicates_exists = df.count() > df.dropDuplicates().count()

dbutils.jobs.taskValues.set(key="has_duplicates", value=duplicates_exists)
```
