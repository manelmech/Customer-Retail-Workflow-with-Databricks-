```python
%sql
CREATE OR REPLACE TABLE ecommerce.retail_customer.customers_orders_silver AS 
SELECT 
c.customer_id,
c.customer_name,
c.state,
c.city,
o.order_number,
o.order_datetime,
o.number_of_line_items,
o.promo_info
FROM ecommerce.retail_customer.customers_bronze c
JOIN ecommerce.retail_customer.orders_bronze o 
ON 
c.customer_id = o.customer_id
```
